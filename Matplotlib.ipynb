```python
# stacked bar chart

import numpy as np

c = ['india','australia','england','new zeland','south africa']
gold = [100,25,50,60,70]
silver = [300,250,400,350,390]
bronze = [500,450,300,250,495]

plt.bar(c,gold,label = 'gold')
plt.bar(c,silver,bottom = gold,label = 'silver')

sum_gold_silver = list(np.add(gold,silver))

plt.bar(c,bronze,bottom = sum_gold_silver,label = 'bronze')
plt.legend()
plt.xlabel('country')
plt.ylabel('medals awarded')
plt.title('asian champinship medls awaded by country')
plt.show()
```
